```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
```

```python
bank=("D:/ExcelRassignments/Assignment6/bank-full.csv")
bank=pd.read_csv(bank,sep=';')
bank.head()
```

```python
bank=bank[['y','age','job','education','default','balance','housing','loan','marital','day','month','duration','campaign','pdays','previous','poutcome','contact']]
```

```python
bank
```

```python
bank['y']=bank['y'].map({'yes':1 , 'no':0})
```

```python
bank['default']=bank['default'].map({'yes':1 , 'no':0})
```

```python
bank['housing']=bank['housing'].map({'yes':1 , 'no':0})
```

```python
bank['loan']=bank['loan'].map({'yes':1 , 'no':0})
```

```python
bank
```

```python
bank1=bank.copy()
```

```python
bank1=bank1.drop('job',axis=1)
bank1 =bank1.drop('education', axis=1)
bank1 =bank1.drop('marital', axis=1)
bank1 =bank1.drop('month', axis=1)       
bank1 =bank1.drop('contact', axis=1)       
bank1 =bank1.drop('poutcome', axis=1)
```

```python
bank1
```

```python
bank1.info()
```

```python
bank1.shape
```

```python
bank1=bank1.dropna()
bank1.shape
```

```python
bank1[bank1.duplicated()]
```

```python
bank1.isna().sum()
```

```python
a = bank1.iloc[:,1:]
b = bank1.iloc[:,0]
```

```python
classifier = LogisticRegression()
classifier.fit(a,b)
```

```python
y_pred = classifier.predict(a)
```

```python
y_pred1 = pd.DataFrame({'actual': b, 'predicted_prob': classifier.predict(a)})
```

```python
y_pred1
```

```python
# Confusion Matrix for Model Accuracy
from sklearn.metrics import confusion_matrix
confusion_matrix= confusion_matrix(b,y_pred)
print(confusion_matrix)
```

```python
classifier.predict_proba(a)
```

```python
# ROC Curve
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score
```

```python
fpr, tpr, thresholds = roc_curve(b, classifier.predict_proba(a)[:,1])
auc = roc_auc_score(b, y_pred)
import matplotlib.pyplot as plt

plt.plot(fpr, tpr, color= 'blue', label = 'Logit Model (area = %0.2f)'%auc)
plt.plot([0,1],[0,1], 'k--')
plt.xlabel('False Positive Rate or [1- True Negative Rate]')
plt.ylabel('True Positive Rate')
```
